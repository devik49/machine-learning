#### Write a Python program to demonstrate EM algorithm using sklearn libraries. Predict the values of latent variables.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

import warnings
warnings.filterwarnings("ignore")
```

```python
# Generate sample data
np.random.seed(0)
samples = np.random.normal(size=(100,2))
```

```python
gmm = GaussianMixture(n_components=2)
gmm.fit(samples)

labels = gmm.predict(samples)

plt.scatter(samples[:,0],samples[:,1],c=labels,cmap='viridis')
plt.show()
```
