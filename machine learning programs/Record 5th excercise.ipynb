```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Automobile_data.csv')
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace = True)
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

enc = LabelEncoder()
df['company'] = enc.fit_transform(df['company'])
df.head()
```

```python
df['body-style'] = df['body-style'].map({'convertible':1,'hatchback':2,'sedan':3,'wagon':4,'hardtop':5})
df['engine-type'].replace({'dohc':1,'ohcv':2,'ohc':3,'l':4,'rotor':5,'ohcf':6,'dohcv':7},inplace = True)
df['num-of-cylinders'].replace({'four':4,'six':6,'five':5,'three':3,'twelve':12,'two':2,'eight':8},inplace = True)
```

```python
x=df.drop(['price'],axis=1)
y=df['price']
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 21)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
dt = DecisionTreeRegressor()
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
print("Decision Tree : ",r2_score(y_test,y_pred)*100)
```

```python
result = pd.DataFrame({'Actual' : y_test, 'Predicted' : y_pred})
result
```

```python
from sklearn.tree import plot_tree
from matplotlib.pylab import rcParams

rcParams['figure.figsize'] = 150,150
plot_tree(dt)
plt.show()
```
