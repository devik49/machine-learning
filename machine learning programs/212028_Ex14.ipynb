#### Write a Python program to demonstrate the application of EM algorithm where the user is supposed to find out values for missing values assuming that it follows Guassian distribution.

```python
import numpy as np
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
# Generate a toy dataset with missing values
np.random.seed(0)
data = np.random.normal(size=(100, 2))
mask = np.random.randint(0, 2, size=(100, 2))
missing = np.ma.masked_array(data, mask)
```

```python
# Fit a Gaussian Mixture Model (GMM) to the observed values
gmm = GaussianMixture(n_components=2)
gmm.fit(missing.compressed().reshape(-1, 1))
```

```python
# Use the EM algorithm to estimate the missing values
X_complete = missing.copy()
n_missing = np.count_nonzero(missing.mask)
for i in range(n_missing):
    x = X_complete[missing.mask][i]
    probs = gmm.predict_proba(x.reshape(1, -1))[0]
    missing_values = gmm.means_.dot(probs.T[0])
    X_complete[missing.mask] = missing_values[1]
```

```python
# Print the estimated missing values
print(X_complete)
```
