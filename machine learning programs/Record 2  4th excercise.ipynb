```python
df1['Species'] = df1['Species'].map({'Iris-setosa':1,'Iris-versicolor':2,'Iris-virginica':3})
df1
```

```python
x = df1.iloc[:, [0,1,2,3]]
```

```python
print(x)
```

```python
y = df1['Species']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler()
x_train =scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
x_train
x_test
from sklearn.decomposition import PCA
```

```python
pca =PCA() 
x_train = pca.fit_transform(x_train)
x_test=pca.transform(x_test)
```

#### 6.Feature Importance

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

```python
features = df1.columns
importances = model.feature_importances_
indices = np.argsort(importances)[-9:]  # top 10 features
plt.title('Feature Importances')
plt.barh(range(len(indices)), importances[indices], color='b', align='center')
plt.yticks(range(len(indices)), [features[i] for i in indices])
plt.xlabel('Relative Importance')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred = dt.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,roc_auc_score,ConfusionMatrixDisplay
```

```python
acc = accuracy_score(y_test, y_pred)*100
cm = confusion_matrix(y_test,y_pred)
print(cm)
cmd = ConfusionMatrixDisplay(cm,display_labels = dt.classes_)
cmd.plot(cmap = 'Blues')
```

```python
import numpy as np
random_state = np.random.RandomState(0)
n_samples, n_features = x.shape
n_classes = len(np.unique(y))
X = np.concatenate([x, random_state.randn(n_samples, 200 * n_features)], axis=1)
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.5, stratify=y, random_state=0)
                     
classifier = DecisionTreeClassifier()
y_score = classifier.fit(X_train, y_train).predict_proba(X_test)
```

```python
from sklearn.preprocessing import LabelBinarizer

label_binarizer = LabelBinarizer().fit(y_train)
y_onehot_test = label_binarizer.transform(y_test)
y_onehot_test.shape  # (n_samples, n_classes)
```

```python
class_of_interest = 3 #vergenica
class_id = np.flatnonzero(label_binarizer.classes_ == class_of_interest)[0]
class_id
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import RocCurveDisplay

RocCurveDisplay.from_predictions(
    y_onehot_test[:, class_id],
    y_score[:, class_id],
    name=f"{class_of_interest} vs the rest",
    color="darkorange",
)
plt.plot([0, 1], [0, 1], "k--", label="chance level (AUC = 0.5)")
plt.axis("square")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("One-vs-Rest ROC curves:\nVirginica vs (Setosa & Versicolor)")
plt.legend()
plt.show()
```
